````python
# Group the dataframe by age and target and count the occurrences of each combination
sex_counts = df.groupby(['sex', 'target']).size().unstack()

# Create a bar plot with two bars for each age value (one bar for target=0, one bar for target=1)
fig, ax = plt.subplots(figsize=(19,5))
ax.bar(sex_counts.index - 0.2, sex_counts[0], width=0.4, color='b', label='no Heart Disease')
ax.bar(sex_counts.index + 0.2, sex_counts[1], width=0.4, color='r', label='Heart Disease')

# Add axis labels and title
ax.set_ylabel('Count')
ax.set_xticks(sex_counts.index)
ax.set_xticklabels(['Female','Male'])
# Add legend
ax.legend()

# Show plot
plt.show()
````

````python
cp_counts =df.groupby(['cp', 'target']).size().unstack()
fig, ax = plt.subplots(figsize=(19,5))
ax.bar(cp_counts.index - 0.2, cp_counts[0], width=0.4, color='b', label='no Heart Disease')
ax.bar(cp_counts.index + 0.2, cp_counts[1], width=0.4, color='r', label='Heart Disease')
ax.set_xticks(cp_counts.index)
ax.set_xticklabels(['typical angina (0)', 'atypical angina(1)', 'non-anginal(2)', 'asymptomatic(3)'])
ax.set_xlabel('chest pain type')
ax.legend()
plt.show()
````

````python
fbs_counts =df.groupby(['fbs', 'target']).size().unstack()
fig,ax=plt.subplots(figsize=(19,5))
ax.bar(fbs_counts.index-0.2,fbs_counts[0],width=0.4, color='b', label='no Heart Disease')
ax.bar(fbs_counts.index+0.2,fbs_counts[1],width=0.4, color='r', label='Heart Disease')
ax.set_xticks(fbs_counts.index)
ax.set_xticklabels(['fasting blood sugar < 120 mg/dl','fasting blood sugar > 120 mg/dl'])
ax.set_xlabel('Fasting blood sugar')
ax.legend()
plt.show()
````

````python
restecg_counts =df.groupby(['restecg', 'target']).size().unstack()
fig,ax=plt.subplots(figsize=(19,5))
ax.bar(restecg_counts.index-0.2,restecg_counts[0],width=0.4, color='b', label='no Heart Disease')
ax.bar(restecg_counts.index+0.2,restecg_counts[1],width=0.4, color='r', label='Heart Disease')
ax.set_xticks(restecg_counts.index)
ax.set_xticklabels(['normal[0]', 'stt abnormality[1]', 'left ventricular hypertrophy[2]'])
ax.set_xlabel('resting electrocardiographic results')
ax.legend()
plt.show()
````

````python
exang_counts =df.groupby(['exang', 'target']).size().unstack()
fig,ax=plt.subplots(figsize=(19,5))
ax.bar(exang_counts.index-0.2,exang_counts[0],width=0.4, color='b', label='no Heart Disease')
ax.bar(exang_counts.index+0.2,exang_counts[1],width=0.4, color='r', label='Heart Disease')
ax.set_xticks(exang_counts.index)
ax.set_xticklabels(['No', 'Yes'])
ax.set_xlabel('exercise-induced angina')
ax.legend()
plt.show()
````

````python
slope_counts =df.groupby(['slope', 'target']).size().unstack()
fig,ax=plt.subplots(figsize=(19,5))
ax.bar(slope_counts.index-0.2,slope_counts[0],width=0.4, color='b', label='no Heart Disease')
ax.bar(slope_counts.index+0.2,slope_counts[1],width=0.4, color='r', label='Heart Disease')
ax.set_xticks(slope_counts.index)
ax.set_xticklabels(['uploping','flat','downsloping'])
ax.set_xlabel('slope of the peak exercise ST segment')
ax.legend()
plt.show()
````

````python
ca_counts =df.groupby(['ca', 'target']).size().unstack()
fig,ax=plt.subplots(figsize=(19,5))
ax.bar(ca_counts.index-0.2,ca_counts[0],width=0.4, color='b', label='no Heart Disease')
ax.bar(ca_counts.index+0.2,ca_counts[1],width=0.4, color='r', label='Heart Disease')
ax.set_xticks(ca_counts.index)
ax.set_xticklabels(['0',' 1','2','3','4'])
ax.set_xlabel('number of major vessels')
ax.legend()
plt.show()
````

````python
df.thal = df.thal.replace({0: 1})
df.thal = pd.to_numeric(df.thal)  # convert thal column to numeric data type
thal_counts =df.groupby(['thal', 'target']).size().unstack()
fig,ax=plt.subplots(figsize=(19,5))
ax.bar(thal_counts.index-0.2,thal_counts[0],width=0.4, color='b', label='no Heart Disease')
ax.bar(thal_counts.index+0.2,thal_counts[1],width=0.4, color='r', label='Heart Disease')
ax.set_xticks([1,2,3])  # set the tick locations
ax.set_xticklabels(['normal (0,1)', 'fixed defect (2)', 'reversable defect (3)'])  # set the tick labels
ax.set_xlabel('thalium stress result')
ax.legend()
plt.show()
````

## Findings:
```
-sex: Males are more likely to have heart disease than Females.
-chest_pain_type: People with chest pain type in typical angina groups are more likely to have heart disease than the other people.
-fasting_blood_sugar: People in 'lower than 120mg/dl' fasting blood suger group are more likely to have heart disease than people in 'greater than 120mg/dl' fasting blood suger group.
-rest_ecg: People in (normal, ST-T wave abnormality) groups are more likely to have heart disease.
-exercise_induced_angina : People in (No for exercice induced angina) group have heart disease more than people in (Yes for exercice induced angina)
-st_slope (the slope of the peak exercise ST segment) : People in (Downslopins: signs of unhealthy heart) a are more likely to have heart disease than the other people.
-thalassemia: {blood disorder} : People in (fixed defect) group are more likely to have heart disease.
```
